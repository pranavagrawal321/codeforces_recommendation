```python
import pandas as pd
import numpy as np
```

```python
def sigmoid(x, slope, midpoint):
    return 1 / (1 + np.exp(-slope * (x - midpoint)))
```

```python
def decayed_sigmoid(x, slope, midpoint, decay_rate, weight):
    return sigmoid(x, slope, midpoint) * np.exp(-decay_rate * x / weight)
```

```python
def label_difficulty(rating, slope, midpoint, decay_rate, weight):
    difficulty = decayed_sigmoid(rating, slope, midpoint, decay_rate, weight)
    if difficulty < 0.33:
        return 'Easy'
    elif difficulty < 0.66:
        return 'Medium'
    else:
        return 'Hard'
```

```python
def segregate_problems(solved_problems, unsolved_problems):
    # Calculate midpoint, slope, decay_rate, and weight based on the rating distribution
    midpoint = solved_problems['rating'].median()
    slope = 0.05  # Adjust the slope based on the sensitivity required
    decay_rate = 0.001  # Adjust the decay rate as needed
    weight = 500  # Adjust the weight as needed

    # Apply difficulty estimation and labeling for solved problems
    solved_problems['difficulty'] = solved_problems['rating'].apply(
        lambda x: label_difficulty(x, slope, midpoint, decay_rate, weight)
    )

    # Apply difficulty estimation and labeling for unsolved problems
    unsolved_problems['difficulty'] = unsolved_problems['rating'].apply(
        lambda x: label_difficulty(x, slope, midpoint, decay_rate, weight)
    )

    return solved_problems, unsolved_problems
```

```python
df_solved_labeled, df_unsolved_labeled = segregate_problems(solved, unsolved)
```
