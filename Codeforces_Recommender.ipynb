```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer()
```

```python
tag_vectors = tfidf_vectorizer.fit_transform(problems['tags'].apply(lambda x: ' '.join(x)))
```

```python
def sigmoid(x, slope, midpoint):
    return 1 / (1 + np.exp(-slope * (x - midpoint)))
```

```python
def decayed_sigmoid(x, slope, midpoint, decay_rate, weight):
    return sigmoid(x, slope, midpoint) * np.exp(-decay_rate * x / weight)
```

```python
def label_difficulty(rating, tags, tag_vectors, tfidf_vectorizer, slope, midpoint, decay_rate, weight):
    difficulty = decayed_sigmoid(rating, slope, midpoint, decay_rate, weight)
    current_tag_vector = tfidf_vectorizer.transform([' '.join(tags)])
    cosine_similarities = (current_tag_vector * tag_vectors.T).toarray().flatten()
    tag_influence = cosine_similarities.sum() / len(cosine_similarities)
    difficulty *= (1 + tag_influence)  # Increase difficulty proportionally to tag influence
    
    if difficulty < 0.33:
        return 'Easy'
    elif difficulty < 0.66:
        return 'Medium'
    else:
        return 'Hard'
```

```python
def segregate_problems(solved_problems, unsolved_problems, tag_vectors, tfidf_vectorizer):
    midpoint = solved_problems['rating'].median()
    slope = 0.05  
    decay_rate = 0.001 
    weight = 500

    solved_problems['difficulty'] = solved_problems.apply(
        lambda row: label_difficulty(row['rating'], row['tags'], tag_vectors, tfidf_vectorizer, slope, midpoint, decay_rate, weight),
        axis=1
    )

    unsolved_problems['difficulty'] = unsolved_problems.apply(
        lambda row: label_difficulty(row['rating'], row['tags'], tag_vectors, tfidf_vectorizer, slope, midpoint, decay_rate, weight),
        axis=1
    )

    return solved_problems, unsolved_problems
```

```python
df_solved_labeled, df_unsolved_labeled = segregate_problems(solved, unsolved, tag_vectors, tfidf_vectorizer)
```
